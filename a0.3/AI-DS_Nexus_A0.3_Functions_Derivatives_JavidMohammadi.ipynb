## 1. Log and Exp: Proving Inverse Relationship


```python
import numpy as np

# Show log(exp(x)) == x and exp(log(x)) == x for several values
x_vals = np.array([1, 2, 10, np.e])
print("x\tlog(exp(x))\texp(log(x))")
for x in x_vals:
    print(f"{x:.2f}\t{np.log(np.exp(x)):.2f}\t\t{np.exp(np.log(x)):.2f}")

# Edge case: try negative and zero values (show error/NaN for log)
y_vals = np.array([0, -1, 5])
print("\nlog(exp(y)) for 0 and negative values:")
print(np.log(np.exp(y_vals)))
```

### üëÄüîé What do you see?

#### **your answer:**
We want to check whether log(exp(x)) is equal to exp(log(x)).

First we created a list of positive numbers.  
Then, we used a loop to test each number in the list and printed the results in columns with two decimal places, using Numpy's exp() and log() functions.  
As the output shows, the identity works well for positive numbers.

Next, we tested zero and negative numbers. However, since the logarithm function cannot take zero or negative values as input, we couldn't evaluate exp(log(x)) for those cases.  

On the other hand, log(exp(x)) is defined for all real numbers, including zero and negatives, and it simply returns x.

## 2. Plotting Sine Functions Variations



```python
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# look at the description of each function and call the proper function from numpy
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.sin(2*x), label='sin(2x)')
plt.plot(x, np.sin(x / 2), label='sin(x/2)')
plt.plot(x, np.sin(x) ** 2, label='sin¬≤(x)')
plt.legend()
plt.title("Variations of Sine Functions")
plt.show()
```

### üëÄüîé What do you see?

#### **your answer:**
The function sin(x) (blue line) starts at (0,0), rises to 1, falls to -1, and then returns to zero, forming a smooth and cyclic wave as x increases.  
The function sin(2x) (orange line) oscillates twice as fast as sin(x), completing more cycles over the same interval.  
In contrast, sin(x/2) (green line) oscillates more slowly, completing only half as many cycles as sin(x).  
The function sin(x**2) (red line) behaves differently -- its oscillations become more rapid as x increase, and it doesn't follow a regular period like the others.

## 4. Features of the Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 200)
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("œÉ(x)")
plt.grid(True)
plt.show()

print("Range:", (y.min(), y.max()), "| At x=0:", sigmoid(0))
```

### üëÄüîé What do you see?

#### **your answer:**
The sigmoid function smoothly rises from values near 0 to values near 1 as x increases.
It most significant change (steepest slope) occurs in the range between -5 and 5.

## 5. Derivatives of Famous Functions

### I have modified some cells.
Let's explore how a function's behavior is reflected in its derivative graph, using some examples.

####Graph 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 400)
plt.plot(x, x**2, label="x¬≤")
plt.plot(x, 2*x, '--', label="d/dx x¬≤")

plt.legend()
plt.title("Functions and Their Derivatives")
plt.show()
```

##### üëÄüîé What do you see?
The function x**2 is a symmetric parabola, curving upward with increasing steepness as x moves away from zero in either direction.  
Its derivative, 2x, is a straight line that passses through the origin.  
At x = 0, the slope is 0 --> the parabola has a minimum point.  
As x increases, the derivative grows --> the funciton is steepening.  

The derivative shows how fast the curve bends -- steeper on both sides, but the direction of the slope (positive or negative) changes with the sign of x.

**In other words:**  
When the **derivative f'(x) = 2x** is **-6**, the **x^2 curve** is **bending quickly downward** -- it's steep and heading into negative values.  
When the derivative is around **-2**, the curve still goes down but **less steeply**.  
On the other side, when the derivative is **+6**, the curve is bending **quickly upward** -- much faster than when it's just **+2**.  

So, **the larger the value (positive or negative) of the derivative**, **the steeper** the x^2 curve is at that point.

####Graph 2

```python
x = np.linspace(0, 3, 400)
plt.plot(x, np.exp(x), label="exp(x)")
plt.plot(x, np.exp(x), '--', label="d/dx exp(x)")

plt.legend()
plt.title("Functions and Their Derivatives")
plt.show()
```

##### üëÄüîé What do you see?
The exponential function grows faster and faster as x increases.  
Its derivative is identical. (d/dx e**x = e^x)

The function's value is its rate of change -- the faster it grows, the faster
it continues to grow.  
This is a unique and fundamental property of the exponential function.

####Graph 3

```python
x = np.linspace(0, 3, 400)

plt.plot(x, np.log(x+0.1), label="log(x)")
plt.plot(x, 1 / (x+0.1), '--', label="d/dx log(x)")

plt.legend()
plt.title("Functions and Their Derivatives")
plt.show()
```

##### üëÄüîé What do you see?

#### **your answer:**
(Here, log(x+0.1) is used to avoid division by zero at x = 0)

The logarithmic function rises **quickly at small values of x**, but it growth **slows down** as x increases.

The drivative 1/x reflects this:  
It is **large** near zero --> rapid change  
It **decreases** toward zero --> the function flattens out

The derivative reveals the **diminishing returns** of logarithmic growth -- rapid early increase follow by slow progress.  
**diminishing returns**: you get less and less benefit (return) from putting in more.

Again, The derivative tells us **how fast** the function is changing.

## 6. Gradient of Selected Functions

```python
# search and learn more about sumpy
import sympy as sp

# Define symbolic variables
x, y = sp.symbols('x y')

# Define the function
f = x**2 * y + sp.sin(y)

# Compute partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("Function: f(x, y) =", f)
print("Partial derivative w.r.t x (‚àÇf/‚àÇx):", df_dx)
print("Partial derivative w.r.t y (‚àÇf/‚àÇy):", df_dy)
```

### üëÄüîé What do you see?

#### **your answer:**
Here, we use the sympy module to calculate the gradient of a function -- that is, the partial derivatives with respect to each variable x and y.

Function:
f(x,y) = x^2*y + sin(y)

Partial derivative with respect to x:  
When differentiating with respect to x, we treat y as a **constant**.  
The derivative of x^2*y is 2xy  
The derivative of sin(y) is 0 (since it contains no x)  

So:
‚àÇf/‚àÇx = 2xy

Partial derivative with respect to y:  
When differentiating with respect to y, we treat x as a constant.  
The derivative of x^2*y is x^2  
The derivative of sin(y) is cos(y)

So:
‚àÇf/‚àÇy = x^2 + cos(y)

    Together, the gradient is:(vector)
    ‚àáf(x,y) = [2xy, x^2 + cos(y)]
